```python
from binary_search_tree import BTree
from movie_collection import MovieCollection, Movie
from cinema_collection import LineCollection, Cinema
from customer import Customer
```

```python
from time import time
from random import random, randint
import matplotlib.pyplot as plt
```

```python
repeates = 1
start = 30_000
finish = 100_000
step = 1_000
colours = ['red', 'blue', 'green']
```

```python
from cProfile import label
from matplotlib.pyplot import title, xlabel, ylabel


fig, ax = plt.subplots(figsize=(10,10))
ax.set(xlabel='Input Size', ylabel='Run time (sec)', title='Empirical Analysis of Btree Algorithms')

algorithm_idx = 0
for al in ['insert', 'search','delete']:
    x = []
    y = []

    for input_size in range(start, finish + step, step):
        ys = []
        
        c0 = Customer(str(random()),str(random()),'','',0)
        for repeat in range(1, repeates + 1):
            root = BTree(c0)
            for z in range(input_size):
                root.insert(Customer(str(random()),str(random()),'','',0))
                
            start_time = time()
            if al == 'insert':
                root.insert(Customer(str(random()),str(random()),'','',0))
            if al == 'search':
                root.search(Customer(str(random()),str(random()),'','',0))
            if al == 'delete':
                root.delete(Customer(str(random()),str(random()),'','',0))

            elapsed_time = time() - start_time

            print(al, input_size, repeat, elapsed_time, sep='\t')

            ys.append(elapsed_time)
        
        ax.scatter([input_size] * repeat, ys, c=colours[algorithm_idx], marker='.')

        y.append(sum(ys) / repeates)
        x.append(input_size)

    ax.plot(x, y, label=al, c=colours[algorithm_idx])
    algorithm_idx += 1

plt.legend()
fig.savefig("result_BTtree.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set(xlabel='Input Size', ylabel='Run time (sec)', title='Empirical Analysis of Movie Collection (STACK) Algorithms')

algorithm_idx = 0
for al in ['insert','display_movie']:
    x = []
    y = []

    for input_size in range(start, finish + step, step):
        ys = []
        
        movies = MovieCollection()
        for repeat in range(1, repeates + 1):
            movie = Movie(str(random()),random())
            for z in range(input_size):
                movies.insert(movie)

            start_time = time()

            if al == 'insert':
                movies.insert(movie)
            if al == 'display_movie':
                movies.display_movie()
            

            elapsed_time = time() - start_time

            print(al, input_size, repeat, elapsed_time, sep='\t')

            ys.append(elapsed_time)
        
        ax.scatter([input_size] * repeat, ys, c=colours[algorithm_idx], marker='.')

        y.append(sum(ys) / repeates)
        x.append(input_size)

    ax.plot(x, y, label=al, c=colours[algorithm_idx])
    algorithm_idx += 1

plt.legend()
fig.savefig("results_movies.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set(xlabel='Input Size', ylabel='Run time (sec)', title='Empirical Analysis of Waiting line customer (QUEUE) Algorithms')

algorithm_idx = 0
for al in ['add_to_line','pick_customer']:
    x = []
    y = []

    for input_size in range(start, finish + step, step):
        ys = []

        for repeat in range(1, repeates + 1):
            customer = Customer(str(random()),str(random()),'','',0)
            waiting_line = LineCollection()
            
            for z in range(input_size):
                waiting_line.add_to_line(Customer(str(random()),str(random()),'','',0)) 

            start_time = time()

            if al == 'add_to_line':
                waiting_line.add_to_line(customer)
            if al == 'pick_customer':
                waiting_line.pick_customer()
            

            elapsed_time = time() - start_time

            print(al, input_size, repeat, elapsed_time, sep='\t')

            ys.append(elapsed_time)
        
        ax.scatter([input_size] * repeat, ys, c=colours[algorithm_idx], marker='.')

        y.append(sum(ys) / repeates)
        x.append(input_size)

    ax.plot(x, y, label=al, c=colours[algorithm_idx])
    algorithm_idx += 1

plt.legend()
fig.savefig("results_waiting_line.png")
plt.show()
```
